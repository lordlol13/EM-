````python
# 1
def is_isomorphic(s, t):
    if len(s) != len(t):
        return False

    map_st = {}
    map_ts = {}

    for c1, c2 in zip(s, t):
        if c1 in map_st and map_st[c1] != c2:
            return False
        if c2 in map_ts and map_ts[c2] != c1:
            return False

        map_st[c1] = c2
        map_ts[c2] = c1

    return True
````

````python
print(is_isomorphic('paper','title'))
````

````python
print(is_isomorphic('paper','tiitle'))
````

````python
# 2
def missing_numbers(nums):
    res = []
    n_max = max(nums)
    s_nums = set(nums)

    for i in range(1, n_max + 1):
        if i not in s_nums:
            res.append(i)
    return res
````

````python
missing_numbers([1, 2, 3, 4, 5, 6, 8, 9, 10, 14])
````

````python
# 3
def prime_factors(n):
    res = []

    while n % 2==0:
        res.append(2)
        n = n/2

    i = 3
    while i*i <= n:
        while n%i == 0:
            res.append(i)
            n = n/i
        i += 2

    if n > 2:
        res.append(int(n))

    return res
````

````python
prime_factors(44)
````
